```python
ano = ano.drop(0)
```

```python
ano['Distância total'] = ano['Distância total'].str.rstrip(' km')
```

```python
ano
```

```python
ano['Distância média'] = ano['Distância média'].str.rstrip(' km')
```

```python
ano
```

```python
ano['Distância máxima'] = ano['Distância máxima'].str.rstrip(' km')
```

```python
ano['Tempo total de atividade'] = ano['Tempo total de atividade'].str.rstrip(' h:m:s')
ano['Tempo em movimento'] = ano['Tempo em movimento'].str.rstrip(' h:m:s')
ano['Tempo médio'] = ano['Tempo médio'].str.rstrip(' h:m:s')
ano['Tempo máximo'] = ano['Tempo máximo'].str.rstrip(' h:m:s')
ano['Subida total'] = ano['Subida total'].str.rstrip(' m')
ano['Média da subida'] = ano['Média da subida'].str.rstrip(' m')
ano['Subida máxima'] = ano['Subida máxima'].str.rstrip(' m')
ano['Descida total'] = ano['Descida total'].str.rstrip(' m')
ano['Média da descida'] = ano['Média da descida'].str.rstrip(' m')
ano['Descida máxima'] = ano['Descida máxima'].str.rstrip(' m')
```

```python
ano
```

```python
ano['Velocidade média'] = ano['Velocidade média'].str.rstrip(' km/h')
ano['Velocidade máxima'] = ano['Velocidade máxima'].str.rstrip(' km/h')
ano['Ritmo médio'] = ano['Ritmo médio'].str.rstrip(' km/h')
ano['Melhor ritmo'] = ano['Melhor ritmo'].str.rstrip(' km/h')
```

```python
ano
```

```python
ano['Frequência cardíaca média'] = ano['Frequência cardíaca média'].str.rstrip(' bpm')
ano['Frequência cardíaca máxima'] = ano['Frequência cardíaca máxima'].str.rstrip(' bpm')
ano['Cadência média de corrida'] = ano['Cadência média de corrida'].str.rstrip(' ppm')
ano['Cadência de corrida máxima'] = ano['Cadência de corrida máxima'].str.rstrip(' ppm')
```

```python
ano
```

```python
ano['Comprimento média de passos'] = ano['Comprimento média de passos'].str.rstrip(' m')
ano['Tempo total'] = ano['Tempo total'].str.rstrip(' h:m:s')
```

```python
ano
```
